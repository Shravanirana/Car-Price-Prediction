```python
X = data.drop(['Car_Name','Selling_Price'],axis=1)
y = data['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,y_train)

xg = XGBRegressor()
xg.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],
             "R2_SCORE":[score1,score2,score3,score4]})
```

```python
import gradio as gr
import pandas as pd
import joblib


model = joblib.load('car_price_predictor')


def predict_car_price(p1, p2, p3, p4, p5, p6, p7):

    fuel_type_map = {'Petrol': 0, 'Diesel': 1, 'CNG': 2}
    seller_type_map = {'Individual': 0, 'Dealer': 1}
    transmission_map = {'Manual': 0, 'Automatic': 1}

    try:
        p3 = fuel_type_map.get(p3, 0)
        p4 = seller_type_map.get(p4, 0)
        p5 = transmission_map.get(p5, 0)


        data_new = pd.DataFrame({
            'Present_Price': [p1],
            'Kms_Driven': [p2],
            'Fuel_Type': [p3],
            'Seller_Type': [p4],
            'Transmission': [p5],
            'Owner': [p6],
            'Age': [p7]
        })


        result = model.predict(data_new)
        return f"Predicted Car Purchase Amount: ₹{round(result[0], 2)}"

    except Exception as e:
        return f"Error: {str(e)}"


iface = gr.Interface(
    fn=predict_car_price,
    inputs=[
        gr.Number(label="Present_Price"),
        gr.Number(label="Kms_Driven"),
        gr.Dropdown(["Petrol", "Diesel", "CNG"], label="Fuel_Type"),
        gr.Dropdown(["Individual", "Dealer"], label="Seller_Type"),
        gr.Dropdown(["Manual", "Automatic"], label="Transmission"),
        gr.Number(label="Owner (0, 1, etc.)"),
        gr.Number(label="Age (in years)")
    ],
    outputs="text",
    title="Car Price Prediction Using Machine Learning"
)


iface.launch()
```
